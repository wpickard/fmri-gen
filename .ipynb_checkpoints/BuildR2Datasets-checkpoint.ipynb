# Build BOLD5000r2 Datasets

Build new datasets using BOLD5000r2 GLM and ROIs from:
 - BOLD5000r1 functional localizers
 - vcAltlas
 - visfAtlas

```python
from pathlib import Path
import re
import h5py
import pandas as pd
from tqdm.auto import tqdm
import nibabel as nib
from nilearn import plotting
import numpy as np
```

```python
dataset_v1 = Path('/s/babbage/b/nobackup/nblancha/public-datasets/bold5000')
dataset_v2 = Path('/s/babbage/b/nobackup/nblancha/public-datasets/bold5000r2')
```

### Utilities

```python
def load_df(participant):
    # Dataframe filename
    df_path = dataset_v2 / f'CSI{participant}_dataframe.pkl'
    
    if df_path.is_file():
        df = pd.read_pickle(df_path)
    else:
        df = None
        
    return df

def get_dataset(image_name):
    if re.match(r'^n[0-9]+', image_name):
        return 'ImageNet'
    elif re.match(r'^COCO', image_name):
        return 'COCO'
    else:
        return 'Scene'
```

## Build Dataframe from image presentation list

```python
for subject in tqdm(range(1,5), desc='Subject'):
    df_path = dataset_v2 / f'CSI{subject}_dataframe.pkl'
    
    # Check for previous versiont of dataframe
    if df_path.is_file():
        if input(f'Overwrite Dataframes for CSI{subject}?[y/n]:') != 'y':
            continue
           
    # Use image presentation lists as base of dataframe
    df = pd.read_csv(dataset_v2 / f'CSI{subject}_imgnames.txt',
                     header=None, 
                     names=['image']
                    )
    # Tag which dataset each image is from
    df['dataset'] = df.image.map(get_dataset)
    df.to_pickle(df_path)
```

## Extract ROIs for each participant

```python
for subject in tqdm(range(1,5), desc='Subject'):
    filename = dataset_v2 / f'CSI{subject}_ROIs.h5'
    
    if filename.is_file():
        if input(f'Overwrite ROIs for CSI{subject}?[y/n]:') != 'y':
            continue
        else:
            filename.unlink()
            
    # Get list of ROI masks using glob from v1 dataset
    mask_list = sorted(list(dataset_v1.glob(f'**/sub-CSI{subject}_mask-*.nii.gz')))
    # Load subject z-score data from v2 dataset
    sub_data = nib.load(dataset_v2 / f'CSI{subject}_GLMbetas-TYPED-FITHRF-GLMDENOISE-RR-ZSCORE-ALL.nii.gz').get_fdata()

    with h5py.File(filename, 'w') as f:
        for mask in tqdm(mask_list, desc='ROI'):
            # Extract ROI name from filename
            roi_name = re.search(r'(?<=_mask-).*(?=\.nii\.gz$)', str(mask)).group(0)
            # Load mask as NIFTI image
            mask = nib.load(mask).get_fdata()
            # Apply mask
            roi = sub_data[mask != 0]
            # Store ROI voxel vector
            f.create_dataset(roi_name, data=roi.T, compression='gzip', compression_opts=9)
```

### Test ROI Shapes

Basic sanity check to ensure v2 ROI vectors are the same shape as the v1.

```python
for subject in range(1, 5):
    print(f'Participant CSI{subject}:')
    with h5py.File(dataset_v2 / f'CSI{subject}_BOLD5000r2.h5', 'r') as v2:
        with h5py.File(dataset_v1 / f'ROIs/CSI{subject}/h5/CSI{subject}_ROIs_TR34.h5', 'r') as v1:
            for key in v1.keys():
                v1_size = v1[key][0].shape[0]
                v2_size = v2[key][0].shape[0]
                print(f'\tKey {key}: v1: {v1_size} v2: {v2_size}', end='\t\t')
                if v1_size == v2_size:
                    print('PASSED')
                else:
                    print('FAILED')
```

## Sanity Check ROI Placement

```python
vcAtlas_img = nib.load('freesurfer/sub-CSI2/mri/lh.Rosenke_vcAtlas.nii.gz')
vcAtlas_data = vcAtlas_img.get_fdata()
csi1_anat = nib.load(dataset_v1 / 'BOLD5000_Structural/CSI2_Structural/T1w_MPRAGE_CSI2.nii')

for index, name in zip(vcAtlas_lut.index, vcAtlas_lut.Name):
    print(f'Index: {index} Name: {name}')
    mask = (vcAtlas_data == index).astype(int)
    plotting.plot_roi(nib.Nifti1Image(mask, vcAtlas_img.affine), csi1_anat, title=name)
```

### Perform for Each Subject

```python
for subject in range(1,5):
    # Check for existing dataset
    filename = dataset_v2 / f'CSI{subject}_vcAtlas_ROIs.h5'
    
    if filename.is_file():
        if input(f'Overwrite ROIs for CSI{subject}?[y/n]:') != 'y':
            continue
        else:
            # Delete existing file
            filename.unlink()

    # We need to load the subject brainmask so we can prevent nans
    brainmask_img = nib.load(dataset_v2 / f'CSI{subject}_mask-BRAIN.nii.gz')
    brainmask_data = brainmask_img.get_fdata()
            
    # Load subject z-score data from v2 dataset
    sub_img = nib.load(dataset_v2 / f'CSI{subject}_GLMbetas-TYPED-FITHRF-GLMDENOISE-RR-ZSCORE-ALL.nii.gz')
    # Sanity check affine alignment
    assert np.array_equal(brainmask_img.affine, sub_img.affine)
    # Load subject data. LONG OPERATION!
    sub_data = sub_img.get_fdata()
    
    # Open new database file
    with h5py.File(filename, 'w') as f:
        for hemi in ['lh', 'rh']:
            # Load Atlas
            atlas_img = nib.load(f'freesurfer/sub-CSI{subject}/mri/{hemi}.Rosenke_vcAtlas.nii.gz')
            atlas_data = atlas_img.get_fdata()
            
            # Sanity check atlas values
            assert np.array_equal(atlas_img.affine, sub_img.affine)
            assert np.array_equal(np.unique(atlas_data), range(0,len(vcAtlas_lut)))            
            
            # Apply brainmask as a supermask on atlas
            atlas_data *= brainmask_data

            for index, name in zip(vcAtlas_lut.index[1:], vcAtlas_lut.Name[1:]):
                # Extract ROI
                roi = sub_data[atlas_data == index]
                # Check for nans
                assert np.count_nonzero(np.isnan(roi)) == 0
                # Create dataset label. We are following the naming convention from BOLD5000
                roi_name = f'{hemi.upper()}{name}'
                # Store ROI voxel vector
                f.create_dataset(roi_name, data=roi.T, compression='gzip', compression_opts=9)
                print(f'Sub: {subject} Hemi: {hemi} Index: {index} Name: {roi_name}')
```

### Test vcAtlas

```python
for subject in range(1, 5):
    print(f'Participant CSI{subject}:')
    brainmask_data = nib.load(dataset_v2 / f'CSI{subject}_mask-BRAIN.nii.gz').get_fdata()
    
    with h5py.File(dataset_v2 / f'CSI{subject}_vcAtlas_ROIs.h5') as f:
        for hemi in ['lh', 'rh']:
            # Load Atlas
            atlas_img = nib.load(f'freesurfer/sub-CSI{subject}/mri/{hemi}.Rosenke_vcAtlas.nii.gz')
            atlas_data = atlas_img.get_fdata()
            atlas_data *= brainmask_data
            
            for index, name in zip(vcAtlas_lut.index[1:], vcAtlas_lut.Name[1:]):              
                roi_name = f'{hemi.upper()}{name}'
                dataset_size = f[roi_name][0].shape[0]
                atlas_size = np.count_nonzero(atlas_data == index)
                
                print(f'\t{roi_name}:')
                size_str = f'\t\tatlas: {atlas_size} dataset: {dataset_size}'
                if atlas_size == dataset_size:
                    result = '\033[32;1mPASSED\033[0m'
                else:
                    result = '\033[31;1mFAILED\033[0m'
                print(f'{size_str:<40}{result}')
                
                num_voxels = f[roi_name].size
                num_nans = np.count_nonzero(np.isnan(f[roi_name][...]))
                nan_str = f'\t\tvoxels: {num_voxels} nans: {num_nans}'
                if num_nans == 0:
                    result = '\033[32;1mPASSED\033[0m'
                else:
                    result = '\033[31;1mFAILED\033[0m'
                print(f'{nan_str:<40}{result}')
```
