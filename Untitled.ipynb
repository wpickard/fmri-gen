```python
def detect_dataset(image_name):
    if re.match(r'^n[0-9]+', image_name):
        return 'ImageNet'
    elif re.match(r'^COCO', image_name):
        return 'COCO'
    else:
        return 'Scene'
```

```python
def get_synset_name(imagenet_name):
    offset = re.search(r'(?<=^n)[0-9]+', imagenet_name).group(0)
    synset = wn.synset_from_pos_and_offset('n', int(offset))
    return synset.name()
```

```python
def get_synset(imagenet_name):
    offset = re.search(r'(?<=^n)[0-9]+', imagenet_name).group(0)
    synset = wn.synset_from_pos_and_offset('n', int(offset))
    return synset
```

```python
def get_ranks(ss):
    paths = ss.hypernym_paths()
    
    path_sets = []
    while True:
        rank = set([path.pop(0) for path in paths if path])
        
        if rank:
            path_sets.append(rank)
        else:
            break

    return path_sets
```

```python
get_ranks(wn.synset('sandbar.n.01'))
```

```python
imagenet_df['ranks'] = imagenet_df.synset.apply(get_ranks)
imagenet_df
```

```python
def reroot(ss, new_root):
    paths = ss.hypernym_paths()
    paths = [path for path in paths if new_root in path]
    
    for path in paths:
        while path[0] != new_root:
            path.pop(0)
    
    return paths
```

```python
reroot(wn.synset('sandbar.n.01'), wn.synset('whole.n.02'))
```
